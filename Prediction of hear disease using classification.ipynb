This code performs the classification of patients having heart disease or not by separaring the predicted values
in the two sets, 0 is for absence and 1 for presence of heart disease where all the predicted values
between 1 and 4 are replaced to 1 to check the model performance

```python
from numpy import genfromtxt
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('TKAgg')
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.decomposition import PCA
import pylab as pl
from itertools import cycle
from sklearn import cross_validation
from sklearn.svm import SVC
```

```python
# Loading and pruning the data
dataset = genfromtxt('Cleveland_data.csv', dtype=float, delimiter = ',')
dataset=np.nan_to_num(dataset)
#print dataset
X = dataset[:,0:12] #Feature set
Y = dataset[:,13] #Label set or target variable
Y
```

```python
# Replacing 1-4 by 1 label
for index, item in enumerate(Y):
    if not (item==0):
        Y[index] = 1
#print(Y)
target_names = ['0','1']
```

```python
# Method to plot graph for reduced dimensions
def plot_2D(data, target, target_names):
    colors = cycle('rgbcykw')
    target_ids = range(len(target_names))
    plt.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        plt.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
        plt.legend()
        plt.savefig('Problem 2 Graph')
```

```python
# Classifying te data using linear SVM and predicting the probability of disease belonging to a particular class
modelSVM = LinearSVC(C=0.1)
pca = PCA(n_components=2, whiten=True).fit(X)
X_new = pca.transform(X)
print(X_new)
#Calling plot_2D
plot_2D(X_new, Y, target_names)
plt.show()
```
